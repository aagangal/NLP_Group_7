```python
import pandas as pd
import nltk
import re
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import numpy as np
import spacy
nltk.download('vader_lexicon')
```

```python


df = pd.read_csv(r'C:\Users\anaya\OneDrive\Desktop\Anay Masters\Intro to NLP\NLP_Group_7\LemmatizedDataset.csv')

analyzer = SentimentIntensityAnalyzer()

print(df.shape[0])

def get_sentiment_score(lyrics):
    sentiment = analyzer.polarity_scores(lyrics)
    pos_sentiment = sentiment['pos']
    neu_sentiment = sentiment['neu']
    neg_sentiment = sentiment['neg']
    if (pos_sentiment > neu_sentiment) and (pos_sentiment > neg_sentiment):
        sentiment_score = 'pos'
    elif (neu_sentiment > pos_sentiment) and (neu_sentiment > neg_sentiment):
        sentiment_score = 'neu'
    elif (neg_sentiment > pos_sentiment) and (neg_sentiment > neu_sentiment):
        sentiment_score = 'neg'
    else:
        sentiment_score = "oof"
    return sentiment_score

sentiment_scorelist = []
for i in range(df.shape[0]):
    sentiment_score = get_sentiment_score(df['Lemmatized_Lyrics'].iloc[i])
    print(i)
    sentiment_scorelist.append(sentiment_score)
df['Sentiment'] = sentiment_scorelist
df.head(11)
```
